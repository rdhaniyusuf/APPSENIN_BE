```python
import requests, json, datetime, time, os
import urllib.parse
start_time = time.time()
```

```python
from dotenv import load_dotenv

print(load_dotenv('.env'))
```

```python
email = os.getenv("USER_NAME_LINK")
password = os.getenv("PASSWORD_LINK")
```

```python
url_login = "https://iofficeplus.iconpln.co.id/v1/backoffice/user/login/v1"
playload_login = {"email":f"{email}","password":f"{password}","type":1}
headers_login = {"Content-Type":"application/json"}
```

```python
def get_login_token(url_login, playload_login, headers_login):
    response = requests.post(url_login, json=playload_login, headers=headers_login)
    data = response.json()
    id = data["data"]["id"]
    name = data["data"]["nama"]
    token = data["data"]["token"]
    return token, id, name
```

```python
token, id, name = get_login_token(url_login, playload_login, headers_login)
```

```python
def get_absensi(token, id):
    month = datetime.datetime.now().month
    year = datetime.datetime.now().year
    url_absensi = "https://iofficeplus.iconpln.co.id/v1/backoffice/absensi/backoffice/kehadiran_detail/v1"
    headers_absensi = {"Authorization": f"Bearer {token}",
    "Content-Type": "application/json"}
    # payload_absensi = {"bulan": month, "idPegawai": f"{id}", "tahun": year}
    payload_absensi = {"bulan": month, "idPegawai": f"26838", "tahun": year}
    response = requests.post(url_absensi, headers=headers_absensi, json=payload_absensi) 
    return response.json()
```

```python
data = get_absensi(token, id)
```
